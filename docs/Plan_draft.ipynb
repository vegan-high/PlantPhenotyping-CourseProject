### Название работы

Улучшение семантической сегментации растений на разных стадиях роста с использованием методов компьютерного зрения
Применение методов компьютерного зрения для цифрового фенотипирования растений на основе RGB изображений

### Описание и обоснование

Цифровое фенотипирование растений — это метод сбора и анализа данных о росте, развитии и физиологическом состоянии растений с использованием цифровых технологий и устройств. Этот подход позволяет исследователям и агрономам более точно и эффективно изучать фенотипические характеристики растений и их реакции на различные условия окружающей среды.

В закрытых системах точная неинвазивная оценка характеристик необходима для:
1. Ускоренная селекция
2. Мониторинг физиологических реакций (исследования)
3. Сравнительный анализ (большие данные)

Цифровое фенотипирование с помощью RGB камер:
1. Дешево 
2. Компактно
3. Универсально

### Исходные данные

- Датасет изображений (около 1 млн) растений на начальных стадиях роста, вид сверху. Ночные и дневные снимки, с промежутками 3-5 минут.
- Кастомный инструмент для сегментации (на основе HSV фильтров)
- Маски для семантической сегментации части датасета
- Модель, обученная на части датасета

### Цели проекта

Несколько целей, от простых к сложным

*Минимум*

1. **Инстанс-сегментация отдельных растений**
Разделение каждого отдельного растения на изображении, когда они не пересекаются.

2. **Сегментация частей растений (листья, стебли)**
Разделение листьев на каждом растении для подсчета их количества и площади.
Разделение стеблей (где применимо, для подсчета длины, геометрии и тп)

*Хорошо бы*

3. **Сегментация перекрывающихся растений**
Выделение отдельных растений, даже если их части перекрываются.

*Максимум*

4. **Использование временной информации**
Включение данных с предыдущих кадров для улучшения сегментации текущего кадра.

5. **Восстановление скрытых частей растений**
Предсказание скрытых частей растений на основании видимых частей и предыдущих данных.

*Конечная цель*

6. Разработка и обучение моделей для 3D реконструкции растений, интеграция с другими сенсорами.
